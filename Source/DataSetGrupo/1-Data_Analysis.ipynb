### Primary Objectives

The primary objectives of this dataset are as follows:

1. **Analysis of Body Signals**: This dataset provides a comprehensive array of physiological measurements, allowing for an in-depth analysis of various body signals and health parameters.

2. **Classification of Smoker or Drinker**: The dataset aims to facilitate the classification of individuals based on their smoking and drinking habits, which can have significant implications for public health and research.

</font>

### Key Attributes in the Dataset

<font face="Times New Roman">

The dataset comprises a variety of attributes that describe the participants' demographic information, health metrics, and lifestyle factors. Here are some of the essential attributes:

- **Sex**: Categorized as male or female.

- **Age**: Rounded up to the nearest 5 years, indicating participants' age.
- **Height**: Rounded up to the nearest 5 cm, representing the participants' height in centimeters.
- **Weight**: Measured in kilograms, indicating the participants' body weight.

- **Systolic Blood Pressure (SBP)**: Measured in millimeters of mercury (mmHg), representing the highest systolic blood pressure.
- **Diastolic Blood Pressure (DBP)**: Measured in mmHg, indicating the diastolic blood pressure.
- **Fasting Blood Glucose (BLDS)**: Measured in milligrams per deciliter (mg/dL), providing information about fasting blood glucose levels.
- **Waist**: Represents the circumference of the individual's waist.

- **Sight Left**: Represents the visual acuity of the individual's left eye.
- **Sight Right**: Represents the visual acuity of the individual's right eye.
- **Hearing Left**: Measures the hearing in the left ear, with 1 representing normal and 2 representing abnormal.
- **Hearing Right**: Measures the hearing in the right ear using the same classification system as hear_left.

- **Total Cholesterol**: Measures the total concentration of cholesterol in mg/dL.
- **HDL Cholesterol**: Measures the concentration of cholesterol in the HDL (high-density lipoprotein) region in mg/dL.
- **LDL Cholesterol**: Measures the concentration of cholesterol in the LDL (low-density lipoprotein) region in mg/dL.
- **Triglyceride**: Measures the total concentration of triglycerides in mg/dL.
- **Hemoglobin**: Measures the concentration of hemoglobin in g/dL.
- **Urine Protein**: Measures the amount of protein in the individual's urine.

- **Serum Creatine**: Measures the concentration of creatinine in the individual's serum in mg/dL.
- **SGOT_AST**: Measures SGOT (Glutamate-oxaloacetate transaminase) - AST (Aspartate transaminase) value in IU/L.
- **SGOT_ALT**: Measures SGOT (Glutamate-oxaloacetate transaminase) - ALT (Alanine transaminase) value in IU/L.
- **Gamma-GTP**: Measures the gamma-GTP value in IU/L, quantifying liver function in the bile duct.

- **Smoking Status**: Represents the individual's smoking state. 1 corresponds to never smoking, 2 corresponds to having used to smoke but quit, and 3 corresponds to still smoking.
- **Drinking Status**: Flag indicating whether the individual is a drinker or not.

<h3 style="font-family: 'Times New Roman'">Initial Enconding</h3>

```python
# Map the values in the "sex" column to numeric values using a dictionary.
# Replace 'Male' with 1 and 'Female' with 0.
df["sex"] = df["sex"].map({'Male': 1, 'Female': 0})
```

```python
# Map the values in the "drinker_or_not" column to numeric values using a dictionary.
# Replace 'Y' with 1 and 'N' with 0.
df["drinker_or_not"] = df["drinker_or_not"].map({'Y': 1, 'N': 0})
```

```python
# Define a dictionary to specify the replacement values. Replace 3 with 2.
smoking_replace = {3: 2}

# Use the replace() method to replace values in the "smoking_state" column.
df["smoking_state"] = df["smoking_state"].replace(smoking_replace)

# Convert the "smoking_state" column to integers
df["smoking_state"] = df["smoking_state"].astype(int)
```
